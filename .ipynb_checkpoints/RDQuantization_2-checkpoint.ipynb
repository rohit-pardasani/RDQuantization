#### Step 11: Load records from RecordsAnno1.csv (22232 Records).<br> Rename PatternType to PatternType1 because we will looking for Type 1 pattern of UHD (Unexpected Hospital Death). <br>Then apply filter to records: Age<=89.0, Sampling Rate =1, No Ventilation, Duration between 24 hrs and 120 hrs. <br>We end up with 2561 records. <br>We save it as RecordsAnno2.csv

```python
import pandas as pd
import numpy as np
anno1 = pd.read_csv('RecordsAnno1.csv',sep=',')
anno1 = anno1.rename(columns = {"PatternType": "PatternType1"})
anno1.head()
anno2 = anno1[(anno1.Age <= 89.0) & (anno1.SamplingRate==1) & (anno1.Duration > 24.0)
              & (anno1.Duration < 120.0) & (anno1.Ventilator==False)].reset_index(drop=True)
display(anno2)
anno2.to_csv('RecordsAnno2.csv')
```

#### Step 13: For downloading numerics of records listed in anno3

```python
import psycopg2
import urllib3
import certifi
from bs4 import BeautifulSoup
import re
import os
import requests
import csv
import requests

import warnings
warnings.filterwarnings('ignore', message='Unverified HTTPS request')

path='./FinalRecords/'
baseurl = 'https://physionet.org/physiobank/database/mimic3wdb/matched/'
proxies = {
  'http': 'http://cis-india-pitc-bangalore.corporate.ge.com:80',
  'https': 'https://cis-india-pitc-bangalore.corporate.ge.com:80',
}
patientIdList = list(set(anno3.PatientId.values))
N = len(patientIdList)
print('Number of patients : ' + str(N))
for i in range(N):
    PatientId = patientIdList[i]
    PatientIdS = 'p'+str(PatientId).zfill(6)
    link = baseurl + PatientIdS[:3] + '/' + PatientIdS + '/'
    #print(link)
    response = requests.get(link, verify=False,proxies=proxies)
    html= BeautifulSoup(response.content.decode('utf-8'))
    for rec in html.find_all('a'):
        record= rec.get('href')
        #print(record)
        pattern= re.compile(".*n.hea")
        pattern1= re.compile(".*n.dat")
        if re.match(pattern,record):
            #print(re.search(r'.*n.hea', record).group())
            match= re.search(r'.*n.hea', record).group()
            link_new= link+ match
            #print(link_new)
            r = requests.get(link_new, verify=False,proxies=proxies)
            soup = BeautifulSoup(r.content, 'html5lib')
            body=soup.body.string
            sub_id= re.search(r'(.*n).dat', body).group(1)
            
            if(int(sub_id[:-1]) in patients_dict[PatientId]):
                filename= sub_id + '.hea'
                body1= re.sub(r'^.*n', sub_id, body)
                body1= body1.encode()
                fl= path+filename
                if(not os.path.isfile(fl)):
                    print('Writing .hea File for patient SNo. '+str(i))
                    with open(fl, "wb") as file:
                        file.write(body1)
                #file.write(r.content)

        if re.match(pattern1,record):
            #print(re.search(r'.*n.dat', record).group())
            match= re.search(r'.*n.dat', record).group(0)
            if int(match[:-5]) in patients_dict[PatientId]:
                link_new= link+ match
                #print(link_new)
                r = requests.get(link_new, verify=False,proxies=proxies)
                fl= path+match
                if(not os.path.isfile(fl)):
                    print('Writing .dat File for patient SNo. '+str(i))
                    with open(fl, "wb") as file:
                        file.write(r.content)
```
