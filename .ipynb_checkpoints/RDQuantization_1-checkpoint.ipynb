#### Step 3: Populate column LastStay using clinical database (don't execute if file RecordsAnno1.csv exists and is complete)

```python
list_patients = anno.PatientId.values
print(len(list_patients))
unique_patients = set(list_patients)
print(len(unique_patients))

# Populate column LastStay
anno.LastStay = False
for patient_id in unique_patients:
    lastrecnum = anno.query('PatientId=='+str(patient_id)).sort_values('DateTimeStamp',ascending=False).iloc[0].loc['RecordNum']
    anno.loc[anno['RecordNum'] == lastrecnum,'LastStay'] = True
print('.....COLUMN LAST STAY POPULATED.....')
anno.head()
```

#### Step 4: Populate column Mortality using clinical database (don't execute if file RecordsAnno1.csv exists and is complete)

```python
conn = psycopg2.connect("dbname=mimic user=postgres  host=/var/run/postgresql/")
cur = conn.cursor()
strquery = "SELECT distinct(SUBJECT_ID) FROM mimiciii.icustays where \
            SUBJECT_ID in (SELECT distinct(SUBJECT_ID) FROM mimiciii.admissions where DISCHARGE_LOCATION='DEAD/EXPIRED') \
            ORDER BY SUBJECT_ID"
cur.execute(strquery)
print("Rows in the data: ", cur.rowcount)
row = cur.fetchone()
mortality_list = []
while row is not None:
    mortality_list.append(row[0])
    row = cur.fetchone()
cur.close()
```

```python
print(mortality_list)
```

```python
anno.Mortality = False
anno.loc[anno['PatientId'].isin(mortality_list),'Mortality']=True
print('.....COLUMN MORTALITY POPULATED.....')
display(anno)
```

#### Step 5: Populate column Gender using clinical database (don't execute if file RecordsAnno1.csv exists and is complete)

```python
conn = psycopg2.connect("dbname=mimic user=postgres  host=/var/run/postgresql/")
cur = conn.cursor()
for patient_id in unique_patients:
    strquery = "SELECT GENDER FROM mimiciii.patients where SUBJECT_ID="+str(patient_id)
    cur.execute(strquery)
    gen = cur.fetchone()[0]
    anno.loc[anno['PatientId'] == patient_id,'Gender'] = gen
cur.close()
print('.....COLUMN GENDER POPULATED.....')
display(anno)
```

#### Step 6: Populate column Age using clinical database and then change ages that are shifted (don't execute if file RecordsAnno1.csv exists and is complete)

```python
conn = psycopg2.connect("dbname=mimic user=postgres  host=/var/run/postgresql/")
cur = conn.cursor()
COUNT = np.shape(anno)[0]
for i in range(COUNT):
    subjectid = anno.iloc[i].loc['PatientId']
    t2 = pd.Timestamp(anno.iloc[i].loc['DateTimeStamp'])
    strquery = "SELECT DOB FROM mimiciii.patients where SUBJECT_ID='"+str(subjectid)+"'"
    cur.execute(strquery)
    dobinfo = cur.fetchone()[0]
    t1 = pd.Timestamp(dobinfo)
    t = t2-t1
    anno.loc[anno['PatientId'] == subjectid,'Age'] = round(t.days/365)
cur.close()
print('.....COLUMN AGFE POPULATED.....')
display(anno)
```

#### Step 7: Populate column HadmId and ICUStayId (don't execute if file RecordsAnno1.csv exists and is complete)

```python
conn = psycopg2.connect("dbname=mimic user=postgres  host=/var/run/postgresql/")
cur = conn.cursor()
COUNT = np.shape(anno)[0]
for i in range(COUNT):
    print(i,end=" ")
    subjectid = anno.iloc[i].loc['PatientId']
    recnum = anno.iloc[i].loc['RecordNum']
    T = pd.Timestamp(anno.iloc[i].loc['DateTimeStamp'])
    strquery = "SELECT SUBJECT_ID,HADM_ID,ICUSTAY_ID,INTIME FROM mimiciii.icustays where SUBJECT_ID="+str(subjectid)
    cur.execute(strquery)
    row = cur.fetchone()
    valueSet = False
    if(row is not None):
        valueSet = True
        hadm_id = row[1]
        icustay_id = row[2]
        intime = pd.Timestamp(row[3])
        t = T-intime
        ts = abs(24*3600*t.days+t.seconds)
    while row is not None:
        new_hadm_id = row[1]
        new_icustay_id = row[2]
        new_intime = pd.Timestamp(row[3])
        new_t = T-new_intime
        new_ts = abs(24*3600*new_t.days+new_t.seconds)
        if(new_ts < ts):
            hadm_id = new_hadm_id
            icustay_id = new_icustay_id
            intime = new_intime
            t = new_t
            ts = new_ts
        row = cur.fetchone()
    if(valueSet):    
        anno.loc[anno['RecordNum'] == recnum,'HadmId'] = int(hadm_id)
        anno.loc[anno['RecordNum'] == recnum,'ICUStayId'] = int(icustay_id)
    
cur.close()
print('.....COLUMN HADMID & ICUSTAYID POPULATED.....')
anno.head()
print(len(anno))
```

#### Step 10: # Populate column ARDS by True or False, depending on whether the patient has a respiratory disease or not. Change the column name to RespDisease instead to ARDS to be more precise (don't execute if file RecordsAnno1.csv exists and is complete)

```python
fa = r'ICD9_mapped_description_updated.csv'
dfa = pd.read_csv(fa)
icd9list = list(dfa.loc[:,'icd9_code'])
icd9list = [str(item) for item in icd9list]
icd9list = str(icd9list)
icd9list = icd9list.replace("[","(")
icd9list = icd9list.replace(']',')')
print(icd9list)
conn = psycopg2.connect("dbname=mimic user=postgres  host=/var/run/postgresql/")
cur = conn.cursor()
strquery = "SELECT distinct(SUBJECT_ID) FROM mimiciii.diagnoses_icd where ICD9_CODE in " + icd9list
cur.execute(strquery)
print("Rows in the data: ", cur.rowcount)
row = cur.fetchone()
ardspatients = []
while row is not None:
    ardspatients.append(int(row[0]))
    row = cur.fetchone()
cur.close()   
print(ardspatients)
```

```python
anno.ARDS = False
anno.loc[anno['PatientId'].isin(ardspatients),'ARDS']=True
anno = anno.rename(columns = {"ARDS": "RespDisease"})
print('.....COLUMN RESP DISEASE POPULATED.....')
display(anno)
```
