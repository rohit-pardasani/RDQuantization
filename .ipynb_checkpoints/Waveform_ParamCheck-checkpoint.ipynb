```python
import psycopg2
import urllib3
import certifi
from bs4 import BeautifulSoup
import re
import os
import requests
import csv
import pandas as pd

import warnings
warnings.filterwarnings('ignore', message='Unverified HTTPS request')

path='./FinalRecords/'
baseurl = 'https://physionet.org/physiobank/database/mimic3wdb/matched/'
proxies = {
  'http': 'http://cis-india-pitc-bangalore.corporate.ge.com:80',
  'https': 'https://cis-india-pitc-bangalore.corporate.ge.com:80',
}

anno1 = pd.read_csv("RecordsAnno1_Mansi.csv")
patientIdList = list(set(anno1["PatientId"]))
N = len(patientIdList)
print('Number of patients : ' + str(N))

recordList = list(set(anno1["RecordNum"]))
print('Number of records : ' + str(len(recordList)))


for i in range(850,anno1.shape[0],1):
    print(i, end=" ")
    paramList=[]
    pid=anno1.loc[i,'PatientId']
    recnum=anno1.loc[i,'RecordNum']
    #display(type(recnum))
    #display(type(pid))
    pids="p"+str(pid).zfill(6)
    link = baseurl + pids[:3] + '/' + pids + '/'
    #print(link)
    try:
        response = requests.get(link, verify=False,proxies=proxies)
        html= BeautifulSoup(response.content.decode('utf-8'))
        #print(html)
        for rec in html.find_all('a'):
            record= rec.get('href')
            pattern= re.compile(str(recnum)+"_[0-9]+.hea")
            if re.match(pattern,record):
                #display("inside")
                match= re.search(str(recnum)+r"_[0-9]+.hea", record).group()
                link_new=link+match
                #display(link_new)
                r=requests.get(link_new,verify=False,proxies=proxies)
                soup = BeautifulSoup(r.content, 'html5lib')
                body=soup.body.string
                lines=body.split("\n")
                for l in lines:
                    #display(l)
                    pattern= re.compile(str(recnum)+"+_[0-9]+.dat")
                    if re.match(pattern,l):
                        #display("inside")
                        param=l.split(" ")[8]
                        #display(param)
                        if param not in paramList:
                            paramList.append(param)

    #display(paramList)
    except Exception as e :
        display("Error with patient: "+ str(pid)+ ",record : " +str(recnum)+ "  :"+e)
    finally:
        anno1.loc[(anno1['PatientId']==pid) & (anno1['RecordNum']==recnum),'WaverformParam']=', '.join(paramList)
        anno1.to_csv("RecordsAnno1_Mansi.csv", index=False)
```
