```python
annofile = r'RecordsAnno3.csv'
dfOuranno = pd.read_csv(annofile, encoding='iso-8859-1') 
d = ['SegmentId','RecordNum','StartingTime','RRBreach','SpO2Breach','RRHighRegions','SpO2LowRegions','Stage1OccurL','Stage2OccurL','Stage3OccurL','RRUnsafeRegions','SpO2UnsafeRegions','Stage1OccurH','Stage2OccurH','Stage3OccurH','FeatureSum','Type1Pattern']
segFeatureFrame = pd.DataFrame(columns = d)
RECORDS = np.shape(dfOuranno)[0]
for i in range(RECORDS):
    if ((dfOuranno.iloc[i]['Category']=='A' or dfOuranno.iloc[i]['Category']=='E') and pd.notnull(dfOuranno.iloc[i]['DTW_RDSegs'])): 
        print(i)
        mystr = dfOuranno.iloc[i]['DTW_RDSegs'].split(';')
        L = len(mystr)
        for j in range(L-1):
            start = float(mystr[j].split(':')[0])
            try:
                addFeatureVectorToSegFrame(i,dfOuranno,start,1,segFeatureFrame)
            except ValueError:
                break
for i in range(RECORDS):
    if ((dfOuranno.iloc[i]['Category']=='C' or dfOuranno.iloc[i]['Category']=='F') and pd.notnull(dfOuranno.iloc[i]['DTW_RDSegs'])): 
        print(i)
        mystr = dfOuranno.iloc[i]['DTW_RDSegs'].split(';')
        L = len(mystr)
        for j in range(L-1):
            start = float(mystr[j].split(':')[0])
            try:
                addFeatureVectorToSegFrame(i,dfOuranno,start,0,segFeatureFrame)
            except ValueError:
                break
```
